```python
import datetime
class Alumno:
    def __init__(self, nombre, dni, fecha_nacimiento):
        self.nombre = nombre
        self.dni = dni
        self.fecha_nacimiento =  fecha_nacimiento
 

class Inscripcion:
    def __init__(self, alumno, fecha_inscripcion):
        self.fechainscripcion= fecha_inscripcion
        self.alumno=alumno
    def datos(self):
        return f"{self.alumno.nombre}, DNI  {self.alumno.dni}, fecha de nacimiento {self.alumno.fecha_nacimiento},  fecha de inscripcion  {self.fechainscripcion}"
 
class Carrera:
    def __init__(self, nombre, inscripciones):
        self.nombre = nombre
        self.inscripciones=inscripciones
    def mostrar_alumnos(self):
        print("Carrera: ", self.nombre )
        aux=0
        print("Alumnos: " )
        for inscripcion in self.inscripciones:
            print(inscripcion.datos())


class Facultad:
    def __init__(self, nombre, carreras):
        self.nombre=nombre
        self.carreras=carreras
    def mostrarCarrerasYAlumnos(self):
        print("Facultad: ", self.nombre)
        for carrera in self.carreras:
            print()
            carrera.mostrar_alumnos()
         


alumno1 = Alumno("Juampi", "11.111.111", datetime.date(1990, 11, 11))
inscripcion1= Inscripcion(alumno1, datetime.date(2000,10,10))
alumno2 = Alumno("Lazza", "22.222.222", datetime.date(1990, 12, 12))
inscripcion2= Inscripcion(alumno2, datetime.date(2000,11,11))

inscripciones=[inscripcion1, inscripcion2]

recursoshidricos = Carrera("Ingeniería en Recursos Hídricos", inscripciones)
informatica = Carrera("Ingeniería en Informática", inscripciones)

carreras=[recursoshidricos, informatica]

facultad = Facultad("FICH", carreras)


facultad.mostrarCarrerasYAlumnos()
```

```python
import hashlib
class Persona:
    def __init__(self, nombre, apellido, clave ):
        self.nombre= nombre
        self.apellido=apellido
        self.clave=hashlib.sha256(clave.encode()).hexdigest()
        
    def validar_password(self, password):
        claveaux= hashlib.sha256(password.encode()).hexdigest()
        return claveaux==self.clave


if __name__ == "__main__":
    persona1= Persona("Rafael", "Gorrochategui","rafacontrasenia")
    print(persona1.validar_password("rafacontrasenia"))
    
```
