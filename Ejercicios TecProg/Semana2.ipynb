```python
import datetime
class Alumno:
    def __init__(self, nombre, dni, fecha_nacimiento):
        self.nombre = nombre
        self.dni = dni
        self.fecha_nacimiento =  fecha_nacimiento
 

class Inscripcion:
    def __init__(self, alumno, fecha_inscripcion):
        self.fechainscripcion= fecha_inscripcion
        self.alumno=alumno
    def datos(self):
        return f"{self.alumno.nombre}, DNI  {self.alumno.dni}, fecha de nacimiento {self.alumno.fecha_nacimiento},  fecha de inscripcion  {self.fechainscripcion}"
 
class Carrera:
    def __init__(self, nombre, inscripciones):
        self.nombre = nombre
        self.inscripciones=inscripciones
    def mostrar_alumnos(self):
        print("Carrera: ", self.nombre )
        aux=0
        print("Alumnos: " )
        for inscripcion in self.inscripciones:
            print(inscripcion.datos())


class Facultad:
    def __init__(self, nombre, carreras):
        self.nombre=nombre
        self.carreras=carreras
    def mostrarCarrerasYAlumnos(self):
        print("Facultad: ", self.nombre)
        for carrera in self.carreras:
            print()
            carrera.mostrar_alumnos()
         


alumno1 = Alumno("Juampi", "11.111.111", datetime.date(1990, 11, 11))
inscripcion1= Inscripcion(alumno1, datetime.date(2000,10,10))
alumno2 = Alumno("Lazza", "22.222.222", datetime.date(1990, 12, 12))
inscripcion2= Inscripcion(alumno2, datetime.date(2000,11,11))

inscripciones=[inscripcion1, inscripcion2]

recursoshidricos = Carrera("Ingeniería en Recursos Hídricos", inscripciones)
informatica = Carrera("Ingeniería en Informática", inscripciones)

carreras=[recursoshidricos, informatica]

facultad = Facultad("FICH", carreras)


facultad.mostrarCarrerasYAlumnos()
```

```python
import hashlib
class Persona:
    def __init__(self, nombre, apellido, clave ):
        self.nombre= nombre
        self.apellido=apellido
        self.clave=hashlib.sha256(clave.encode()).hexdigest()
        
    def validar_password(self, password):
        claveaux= hashlib.sha256(password.encode()).hexdigest()
        return claveaux==self.clave


if __name__ == "__main__":
    persona1= Persona("Rafael", "Gorrochategui","rafacontrasenia")
    print(persona1.validar_password("rafacontrasenia"))
    
```

```python
class Vehiculo:
    def __init__(self, marca, modelo, patente, precio, kilometraje, estado):
        pass

class Autos(Vehiculo):
    def __init__(self, marca, modelo, patente, precio, kilometraje, estado, adicional):
        self.marca=marca
        self.modelo=modelo
        self.patente=patente
        self.precio=precio
        self.kilometraje=kilometraje
        self.estado=estado
        self.adicional=adicional
    def cualEs(self):
        return f"auto"
    def verEstado(self):
        return str(self.estado)
    def verAdicional(self):
        return str(self.adicional)
    def verValor(self):
        self.precio=self.precio+self.adicional.impuesto
        return self.precio

class Motocicleta(Vehiculo):
    def __init__(self, marca, modelo, patente, precio, kilometraje, estado):
        self.marca=marca
        self.modelo=modelo
        self.patente=patente
        self.precio=precio
        self.kilometraje=kilometraje
        self.estado=estado
    def cualEs(self):
        return f"moto"


class Camioneta(Vehiculo):
    def __init__(self, marca, modelo, patente, precio, kilometraje, estado):
        self.marca=marca
        self.modelo=modelo
        self.patente=patente
        self.precio=precio
        self.kilometraje=kilometraje
        self.estado=estado
    def cualEs(self):
        return f"camioneta"
    

class EstadoVehiculo:
    def __init__(self, Cliente):
        pass
    def __str__(self):
        pass
        

class Usado(EstadoVehiculo):
    def __init__(self, Cliente):
        self.cliente=Cliente
    def __str__(self):
        return f"Usado"
class Nuevo(EstadoVehiculo):
    def __init__(self):
        pass
    def __str__(self):
        return f"Nuevo"
class Cliente:
    def __init__(self, nombre, apellido, telefono):
        self.nombre=nombre
        self.apellido=apellido
        self.telefono=telefono

class Adicional:
    def __init__(self, pais, impuesto):
        pass
    def __str__(self):
        pass
class Importado(Adicional):
    def __init__(self, pais, impuesto):
        self.pais=pais
        self.impuesto=impuesto
    def __str__(self):
        return f"Importado"
class Nacional(Adicional):
    def __init__(self, pais, impuesto):
        self.impuesto=0
        pass
    def __str__(self):
        return f"Nacional"
    
class Venta:
    def __init__(self, vehiculo, fecha, Cliente):
        self.vehiculo=vehiculo
        self.fecha=fecha
        self.cliente=Cliente

class Empresa:
    def __init__(self):
        self.venta=[]
    def añadirVenta(self, venta):
        self.venta.append(venta)
    def totalAutosNacionalesUsados(self):
        aux=0
        for venta in self.venta:
            
            if venta.vehiculo.cualEs()=="auto": 
                if venta.vehiculo.verEstado()=="Usado" and venta.vehiculo.verAdicional()=="Nacional":
                    aux+=venta.vehiculo.verValor()
        return aux

   
    
duenio = Cliente("Juan", "Perez", "123456")

# Estado usado
estado_usado = Usado(duenio)

# Auto nacional
nacional = Nacional("Argentina", 0)

# Auto
auto1 = Autos("Ford", 2015, "ABC123", 10000, 80000, estado_usado, nacional)

# Cliente comprador
comprador = Cliente("Carlos", "Lopez", "999999")

# Venta
venta1 = Venta(auto1, "2026-03-28", comprador)

# Empresa
empresa = Empresa()
empresa.añadirVenta(venta1)

# Probar función
print("Total:", empresa.totalAutosNacionalesUsados())
```
