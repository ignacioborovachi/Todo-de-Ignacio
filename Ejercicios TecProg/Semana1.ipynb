Ejercicio 2: Realiza un programa que convierta un número de segundos introducidos por el usuario en horas,
minutos y segundos.


```python
seg = int(input("Cantidad de segundos: "))
horas= int(seg/3600)
minutos = int((seg%3600)/60)
segundos = int(seg%60)
print ("Esa cantidad de segundos equivale a: ", horas, ":", minutos, ";", segundos)
```

Ejercicio 2: Crea un programa que calcule y muestre la raíz cuadrada de una serie de números introducidos por
el usuario. El programa termina cuando el usuario introduce un número negativo.


```python
import math
n = int(input("Cantidad de numeros: "))
for i in range(n):
    num = int(input("Numero: "))
    if num<0:
        break
    print("Su raiz es: ", math.sqrt(num))
 
```

Ejercicio 4: 
Crea un programa que solicite la cantidad de kilovatios hora (kWh) consumidos en un mes y calcule
el costo total del servicio según la siguiente tabla:

● Bajo. Hasta 100 kWh → $200 por kWh

● Medio. Entre 101 y 300 kWh → $300 por kWh

● Alto. Más de 300 kWh → $400 por kWh

El programa debe mostrar el total a pagar y la categoría de consumo.

```python
kwh =int(input("Total kilovatios hora"))
if kwh<=100:
    pagar=kwh*200
    cat="baja"
elif kwh <=300:
    pagar = kwh*300
    cat="media"
elif kwh >300:
    pagar = kwh*400
    cat="alta"
print("Total a pagar: ",  pagar, "y pertenece a la categoria ", cat)
```

Ejercicio 6: Escribe un programa que calcule la suma de todos los números pares hasta un número N.

Ejercicio 7: Desarrolla un programa que cuente el número de dígitos de un número entero N.

Ejercicio 8: Escribe un programa que convierta un número entero N a binario.

```python
n= int(input("Numero para pasarlo a binario: "))
binario = ""
while n>0:
    if n%2==0: 
        binario= "0"+binario
    else:
        binario="1"+binario
    n=n//2
print(binario) 
```
